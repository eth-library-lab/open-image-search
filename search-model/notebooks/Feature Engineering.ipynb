```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
from ast import literal_eval
```

```python
metadata_path = "../data/processed/ethz/metadata/metadata.csv"
df_meta = pd.read_csv(metadata_path, index_col='id')
print(len(df_meta))
df_meta.head()
```

```python
def make_one_hot_feature_df(df_meta, col_name, is_nested=False):

    if is_nested:
        df_id = df_meta[[col_name]]
        df_id = df_id.fillna('[]')
        df_id[col_name] = df_id[col_name].apply(lambda x: literal_eval(x))
        ser_id = df_id[[col_name]].explode(col_name)
    else:
        ser_id = df_meta[col_name] 
    
    df_id = pd.get_dummies(ser_id, prefix=col_name).groupby(level=0).sum()
    
    return df_id
```

```python
df_inner = pd.merge(df_feat, df_meta, on='fpath', how='inner')
print(len(df_inner))
df_inner.head()
```

```python
df_left = pd.merge(df_feat, df_meta, on='fpath', how='left')
print(len(df_left))
df_left.head()
```

```python
# 43 prints without metadata
pd.concat([df_inner,df_left]).drop_duplicates(keep=False)
```

```python
from sklearn.preprocessing import OneHotEncoder

oe_style = OneHotEncoder()
oe_results = oe_style.fit_transform(df_left[["classification"]])
class_df = pd.DataFrame(oe_results.toarray(), columns=oe_style.categories_)
# drop the last NaN column
class_df = class_df.drop(class_df.columns[[-1]], axis=1)
class_df.head()
```

```python
df_left = df_left.join(class_df)
print(len(df_left))
df_left.head()
```

```python
# Fit the multi-label binarizer on material_technique
mlb = MultiLabelBinarizer()
mat_tec_series = df_inner['material_technique'].apply(lambda s: [l for l in str(s).split(',')])
mlb.fit(mat_tec_series)
material_technique_df = pd.DataFrame(mlb.transform(mat_tec_series), columns=mlb.classes_)
```

```python
df_left = df_left.join(material_technique_df)
# remove the Nan column from material_technique classes
df_left = df_left.drop(columns = ['nan'])
print(len(df_left))
df_left.head()
```
