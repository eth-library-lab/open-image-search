```python
tdf = df.dropna(subset=['relations'])
tdf = tdf.set_index('recordID')
```

```python
ser_rel = tdf['relations'].dropna()
# ser_rel = ser_rel[:20]
ser_rel = ser_rel.str.split(";")
```

```python
ser_rel.head()
```

```python
relations_list = []
for i, row in ser_rel.iteritems():

    row_relations = [(i,) + tuple(el.split(',', maxsplit=1)) for el in row]
    relations_list.extend(row_relations)


df_rel = pd.DataFrame(data=relations_list, columns=['recordID','relation_type','relation_invNr'])
df_rel['relation_type'] = df_rel['relation_type'].str.strip()
df_rel['relation_invNr'] = df_rel['relation_invNr'].str.strip()
```

```python
df_rel.head()
```

```python
relations_list
```

```python
df_rel[df_rel['recordID'].isin([13245,28840,135484])]
```

```python
record_id_dict = dict(zip(df['invNr'].to_list(), df['recordID'].to_list()))
```

```python
# drop na relation_invNr
df_rel = df_rel.dropna(subset=['relation_invNr'])
```

```python
df_rel.loc[:,'relation_record_id'] = df_rel['relation_invNr'].map(record_id_dict)
df_rel
```

```python
# records with invNr that couldn't be mapped to a recordId
fltr = df_rel['relation_record_id'].isna()
df_rel.loc[fltr,:]
```

```python
df_rel = df_rel.loc[~fltr,:].sort_values(by=['relation_type'])
df_rel['relation_record_id'] = df_rel['relation_record_id'].astype(int)
```

```python
df_rel[df_rel['relation_record_id'].isin([13245,28840,135484])]
```

```python
relation_types = list(df_rel['relation_type'].unique())
relation_types
```

```python
df_rel['relation_type'].value_counts()
```

```python
# only use certain relationships (to be confirmed)

types_to_keep = ['Zustandsvariante','Abzug', 'Dublette'] # ['Variante','Teil von','Kopie', 'Original zu Kopie', 'Probedruck']
fltr = df_rel['relation_type'].isin(types_to_keep)
df_rel = df_rel.loc[fltr, :]
df_rel
```

```python
# filter out duplicates
df_rel_ids = df_rel[['recordID','relation_record_id']].drop_duplicates()
```

```python
df_rel[df_rel['relation_record_id'] == 135484]#
```

```python

record_ids_sorted = df_rel_ids.groupby('recordID').count().sort_values(by='relation_record_id', ascending=False)
record_ids_sorted = record_ids_sorted.rename(columns={"relation_record_id":"related_records_count"})
```

```python
record_ids_sorted
```

```python
related_records_list = []
for rec_id in record_ids_sorted.index:
    fltr = df_rel_ids['recordID'] == rec_id
    rec_lst = np.sort(df_rel_ids.loc[fltr,'relation_record_id'].values).tolist()
    related_records_list.append(rec_lst)
```

```python
record_ids_sorted['related_records'] = related_records_list
```
